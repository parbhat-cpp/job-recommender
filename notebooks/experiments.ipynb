```python
df = pd.read_csv('datasets/jobs_dataset.csv')
df
```

```python
df.shape
```

```python
df.drop(columns=['searchInput/position', 'searchInput/country', 'jobType/1', 'jobType/2', 'jobType/3'], inplace=True)
```

```python
df.head()
```

```python
df['jobType/0'].fillna(df['jobType/0'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.rename(columns={"jobType/0": "job_type"}, inplace=True)
```

```python
df['tags'] = df['positionName'] + df['location'] + df['description'] + df['job_type']
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
df
```

```python
from gensim.utils import simple_preprocess
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    tokens = simple_preprocess(text, deacc=True, min_len=3)
    
    clean_tokens = [
        lemmatizer.lemmatize(word) for word in tokens if word not in stop_words
    ]
    return clean_tokens
```

```python
tags = df['tags'].apply(preprocess_text)
```

```python
len(tags),df.shape
```

```python
X = tags.apply(lambda x: " ".join(x))
y = df.index.tolist()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape,X_test.shape,len(y_train),len(y_test)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_train_vector = vectorizer.fit_transform(X_train)
X_test_vector = vectorizer.transform(X_test)
```

```python
from src.utils import load_numpy_array

vec = load_numpy_array('../artifacts/07_30_2025_19_47_10/data_transformation/transformed/train.npy')
print(vec)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_input = "I am a fresherlooking for an analyst job in Canada"
input_vector = vectorizer.transform([user_input])

similarity_score = cosine_similarity(input_vector, X_train_vector).flatten()
top_indices = similarity_score.argsort()[-5:][::-1]

output_list = []

for i in top_indices:
    output_list.append(df.iloc[i])

output_df = pd.DataFrame(output_list, columns=['company', 'rating', 'location', 'positionName', 'description', 'salary', 'url', 'job_type', 'externalApplyLink'])
output_df
```
